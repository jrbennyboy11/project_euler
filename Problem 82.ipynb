```python
import pandas as pd
import numpy as np
from multiprocessing import Pool
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
pd.options.display.max_rows    = 1000
pd.options.display.max_columns = 1000

'''
matrix = [
    [131, 673, 234, 103, 18 ],
    [201, 96 , 342, 965, 150],
    [630, 803, 746, 422, 111],
    [537, 699, 497, 121, 956],
    [805, 732, 524, 37 , 331]
]
'''

matrix_file = "p082_matrix.txt"
matrix = []

with open(matrix_file,'r') as f:
    for line in f:
        matrix.append([int(val.strip()) for val in line.split(',')])

class MatrixSolver:
    def __init__(self,matrix):
        self.matrix = matrix
        self.min = sum(matrix[0])
        self.min_arr = matrix[0]
        self.m_len = len(matrix)
    
    def solve_helper(self,y,x=0,total=0,total_arr=[]):
        grand_total = total + self.matrix[y][x]
        arr = total_arr + [self.matrix[y][x]]
        
        if grand_total > self.min:
            return
        
        if x == (self.m_len - 1):
            if grand_total < self.min:
                self.min = grand_total
                self.min_arr = arr
                print("New Min: ",grand_total,arr)
            else:
                return
        else:
            self.solve_helper(y,x+1,grand_total,arr)
            if y > 0:
                self.solve_helper(y-1,x,grand_total,arr)
            if y < (self.m_len-1):
                self.solve_helper(y+1,x,grand_total,arr)
    
    def solve(self):
        p = Pool(8)
        p.map(self.solve_helper,[y for y in range(0,self.m_len)])
        
        '''
        for start in range(0,self.m_len):
            y = start
            x = 0
            self.solve_helper(0,[],x,y)
        
        '''
        
        print("Finished Solving Matrix.")
        print(" ## Solution ## ")
        print(self.min,":",self.min_arr)
        
```

```python
project_euler_problem_46 = MatrixSolver(matrix)
project_euler_problem_46.solve()
```
